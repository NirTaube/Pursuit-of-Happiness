```python
file_path = 'data.csv'

# Read the CSV file
data = pd.read_csv(file_path)

# Display the data
data.head()
```

```python
data.columns
```

```python
data_cleaned = data.drop(['Series Code','Country Code','1990 [YR1990]', '2000 [YR2000]', '2013 [YR2013]', '2014 [YR2014]',
       '2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]'],axis=1)
```

```python
data_cleaned.head()
```

```python
data_cleaned = data_cleaned.rename(columns={'2018 [YR2018]': '2018', '2019 [YR2019]': '2019', '2020 [YR2020]': '2020', '2021 [YR2021]': '2021', '2022 [YR2022]': '2022'})
```
